Tạo list để lưu các giá trị sai biệt


```python
columns_to_diff = ['cases_new', 'deaths_new', 'tests_total', 'cases_active', 'cases_recovered']
max_continent = []
for c in df['continent'].unique():
    continent_data = df[df['continent'] == c]
    continent_diff = continent_data[columns_to_diff].diff().max().max()
    max_continent.append(continent_diff)

max_continent_dict = dict(zip(df['continent'].unique(), max_continent)) # Use df instead of d
df['max_continent'] = df['continent'].map(max_continent_dict) # Assign to df
print(df.head())
continent_data = pd.DataFrame(df['max_continent'])
print(continent_data.head(5))
d= df.drop(['max_continent'], axis=1)
```
